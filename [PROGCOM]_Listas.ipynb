##Trabajar y acceder listas
1. Si se tiene una lista a con 3 elementos almacenados en ella, se puede acceder a los mismos usando corchetes y un índice, que va desde 0 a n-1 siendo n el tamaño de la lista.

````python
a = [90, "Python", 3.87]
for n in range(len(a)):
  print(f"Este es el elemento {n+1} de la lista {a[n]}")

  #Recordar que python empieza a contar en la posición 0!!
````

````python
#accediendo a posiciones negativas:
print(a[-1])
print(a[-2])
print(a[-3])
````

También se puede modificar elementos la lista:

````python
a[2]=1
print(a)
````

pero se pueden modificar múltiples elementos

````python
l = [1, 2, 3, 4, 5, 6]
l[0:3] = [0, 0, 0]#Se modifica la posición 0 a la 2
print(l)
````

Y puede ser eliminado

````python
#Para este ejercicio, se eliminará la segunda posición de la lista
l=list(range(1,6))
del l[1]
print(l)
````

Una lista puede contener diferentes listas (**listas anidadas**). Para acceder a sus elementos, se debe usar [] tantas veces como niveles de anidado se tengan

````python
x=[1,2,3,["banano","sandía","mora",["verde","roja"]]]
print(x[3][0])#Se accede a la primera posición de la lista de la cuarta posición general
print(x[1]) #Se accede al primer elemento de la lista
````

También se puede crear sublistas de una lista principal:

````python
nueva1=l[0:2]
nueva2=l[1:4]
print(f"l={l}")
print(f"nueva1={nueva1}")
print(f"nueva2={nueva2}")
````

Con el operador **+** se puede agregar nuevos elementos a la lista

````python
lista = [1, 2, 3]
lista += [4, 5]
print(lista)
````

##ITERANDO LISTAS
Empleando estructuras de repetición, se puede recorrer (**iterar**) los diferentes elementos de la lista

````python
lista1=[5,6,7]
for l in lista1:
  print(l)
````

Si de forma adicional, se necesita imprimir la lista y sus index, puede resolverse así

````python
for index,l in enumerate(lista1):
  print(index,l)
````

Pero también es posile iterar dos listas a la vez:

````python
cod=["U0012345","U0012678","U0023451"]
estudiante=["Ana","Javier","Raquel"]
for c,e in zip(cod,estudiante):
  print(c,e)
````
